A broad class of second order linear homogeneous differential equations can, with some manipulation, be put into the form (Sturm-Liouville)

\

$(p(x)u^{\prime})^{\prime}+q(x)u = \lambda w(x)u$

Assume that the functions p, q, and w are real, and use manipulations much like those that led to the identity Eq. (5.15). Derive the analogous identity for this new differential equation. When you use separation of variables on equations involving the Laplacian you will commonly come to an ordinary differential equation of exactly this form. The precise details will depend on the coordinate system you are using as well as other aspects of the PDE.

\

\

Dos soluciones que satisfacen el problema de Sturm-Liouville son

\

$(p(x)u^{\prime}_{1})^{\prime} + q(x)u_{1} = \lambda_{1}w(x)u_{1} \hspace{2em} \cdots\cdots (1)$

\

$(p(x)u^{\prime}_{2})^{\prime} + q(x)u_{2} = \lambda_{2}w(x)u_{2} \hspace{2em} \cdots\cdots (2)$


\

\

Ahora se proce a multiplicar $u_{1}^{*}$ por la ecuacion (2) y $u_{2}^{*}$ por la ecuacion (1), obteniendo

$u^{*}_{2}\left((p(x)u^{\prime}_{1})^{\prime} + q(x)u_{1}\right) = \lambda_{1}w(x)u_{1}u^{*}_{2} \hspace{2em} \cdots\cdots (3)$

\

$u_{1}(p(x)u^{\prime *}_{2})^{\prime} + q(x)u_{1}u^{*}_{2} = \lambda_{2}w(x)u_{1}u^{*}_{2} \hspace{2em} \cdots\cdots (4)$

\

\

Restando las ecuaciones (3) y (4) se obtiene

$u^{*}_{2}\left((p(x)u^{\prime}_{1})^{\prime} + q(x)u_{1}\right) - u_{1}(p(x)u^{\prime *}_{2})^{\prime} + q(x)u_{1}u^{*}_{2} = u^{*}_{2}\lambda_{1}w(x)u_{1} - \lambda_{2}w(x)u_{1}u^{*}_{2}$

\

$u_{2}^{*}(p(x)u^{\prime}_{1}) - u_{1}(p(x)u^{\prime *}_{2})^{\prime} = u_{2}^{*}(\lambda_{1}w(x)u_{1}) - u_{1}\lambda_{2}w(x)u_{2}^{*}$

\

$u^{*}_{2}(p(x)u_{1}^{\prime})^{\prime} - u_{1}(p(x)u^{\prime *}_{2})^{\prime} = (\lambda_{1} - \lambda_{2})w(x)u_{1}u_{2}^{*}$

\


$\int_{a}^{b} u^{*}_{2}(p(x)u_{1}^{\prime})^{\prime} - u_{1}(p(x)u^{\prime *}_{2})^{\prime} dx = \int_{a}^{b} (\lambda_{1} - \lambda_{2})w(x)u_{1}u_{2}^{*} dx$

\

#Ejercicio 2: 10.23 en la lectura

Este problema hace referencia a unas ecuaciones en la [lectura](http://www.physics.miami.edu/~nearing/mathmethods/ode.pdf) de la sección de ecuaciones diferenciales ordinarias.

The Laplacian in cylindrical coordinates is Eq. (10.49). Separate variables for the equation $∇^{2}V =0$ and you will see that the equations in $z$ and $\phi$ are familiar. The equation in the r variable is less so, but you’ve seen it (almost) in Eqs. (4.18) and (4.22). Make a change of variables in the $r$-differential equation, $r = kr$, and turn it into exactly the form described there.


\

Sabemos que la ecuacion de Laplace esta dada por la siguiente expresion

$$\nabla^{2}V = 0$$

\

Por lo tanto, la ecuacion de Laplace en cordenadas cilindricas es

$$\frac{1}{r}\frac{\partial}{\partial r}\left(r\frac{\partial V}{\partial r}\right) + \frac{1}{r^{2}}\frac{\partial^{2} V}{\partial\theta^{2}}+\frac{\partial^{2}V}{\partial z^{2}} = 0 \hspace{2em}\cdots\cdots (1) $$

\

Con el metodo de separacion de variables decimos que $V(r, \theta, z) = R(r)Q(\theta)Z(z)$

Sustituyendo se obtiene lo siguiente:


 $$\nabla^2 V = \frac{1}{r}\frac{\partial }{\partial r}\left(r \frac{\partial R}{\partial r}\right)QZ+\frac{1}{r^2}\frac{\partial^2 Q}{\partial \theta^2}R Z+\frac{\partial^2 Z}{\partial z^2}RQ$$

\

\

Dividiendo la ecuacion (1) entre $V(r,\theta, z)=R(r)Q(\theta)Z(z)$ se obtiene:

$$\frac{1}{r R}\frac{\partial }{\partial r}\left(r \frac{\partial R}{\partial r}\right)+\frac{1}{r^2 Q}\frac{\partial^2 Q}{\partial \theta^2}+\frac{1}{Z}\frac{\partial^2 Z}{\partial z^2}=0$$

\

\

Note que podemos reagrupar cada termino de la siguiente manera:

$$\left[\frac{1}{r R}\frac{\partial}{\partial r}\left(r \frac{\partial R}{\partial r}\right)+\frac{1}{r^2 Q}\frac{\partial^2 Q}{\partial \theta^2}\right]+\frac{1}{Z}\frac{\partial^2 Z}{\partial z^2}=0 \hspace{2em}\cdots\cdots(2)$$

\


\

Para que la ecuacion (2) se cumpla que la ecuacion debe de ser igual a cero. Por la dependencia de los terminos respecto a sus dichas variables independiente, solo se cumple si:

$$\frac{1}{r R}\frac{\partial}{\partial r}\left(r \frac{\partial R}{\partial r}\right)+\frac{1}{r^2}\frac{\partial^2 Q}{\partial \theta^2}= -l^2 \hspace{2em}\cdots\cdots (3)$$

$$Z\frac{\partial^2 Z}{\partial z^2}=l^2Z \hspace{10em}\cdots\cdots (4)$$


Multiplicando la ecuacion (3) por $r^2$ tenemos

$$\frac{r}{R}\frac{\partial}{\partial r}\left(r\frac{\partial R}{\partial r}\right)+\frac{1}{Q}\frac{\partial^2 Q}{\partial \theta^2}= -l^2r^2$$


\

\

Nuevamente reagrupando los terminos:

$$\left[\frac{r}{ R}\frac{\partial }{\partial r}\left(r \frac{\partial R}{\partial r}\right)+l^2r^2\right]+\left[\frac{1}{Q}\frac{\partial^2 Q}{\partial \theta^2}\right]=0$$

Esto se cumple si,


$$\left[\frac{r}{R}\frac{\partial }{\partial r}\left(r \frac{\partial R}{\partial r}\right)+l^2r^2\right]=k^2 \hspace{4em} \cdots\cdots (5)$$


$$\left[\frac{1}{Q}\frac{\partial^2 Q}{\partial \theta^2}\right]=-k^2 \hspace{9em} \cdots\cdots (6)$$



Multiplicando las ecuaciones (5) y (6) por $R$ y $Q$ obtenemos



$$\left[ r\frac{\partial }{\partial r}\left(r \frac{\partial R}{\partial r}\right)+l^2r^2R\right]=k^2R$$

$$\left[\frac{\partial^2 Q}{\partial \theta^2}\right]=-k^2Q$$

\

\

Como se puede ver la ecuacion diferencial parcial a 3 ecuaciones diferenciales parciales.

$$r\frac{\partial }{\partial r}\left(r \frac{\partial R}{\partial r}\right)+[l^2r^2 -k^2]R=0$$

$$\left[\frac{\partial^2 Q}{\partial \theta^2}\right]=-k^2Q$$

$$Z\frac{\partial^2 Z}{\partial z^2}=l^2Z$$

Las soluciones a estas ecuaciones diferenciales son:

Las funciones de Bessel:

$$\boxed{R_{k}(r)=A_0 J_k(lr)+B_0N_k(lr)}$$

\

Las funciones trigonometricas:

$$\boxed{Q_{k}(\theta)= A_1sin(k \theta)+B_2 cos(k \theta)}$$


\

Finalmente,

$$\boxed{Z_{l}(z)=A_2 e^{l z}+B_2 e^{-l z}} $$

\

\


Se sabe que para el caso en que $l=0$, las soluciones van a ser

$$Z_{0}(z)=Cz+D$$


Y al sustituir $r'=lr$, se obtiene

$$\frac{\partial}{\partial r}=\frac{\partial}{\partial r'}\frac{\partial r'}{\partial r}=\frac{\partial}{\partial r'}l$$

Sustituyendo esto en la ecuacion radial finalmente se obtiene


$$\boxed{r'\frac{\partial }{\partial r'}(r' \frac{\partial R}{\partial r'})+[r^{'2} - k^2]R=0}$$

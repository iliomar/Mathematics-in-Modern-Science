Utilizando como base $\beta \lbrace 1, x, x^{2}, x^{3}, ... \rbrace$.

\

Tenemos que los polinomios de Legendre son los siguientes

\

$\hspace{6em}\boxed{P_{0}(x) = 1}$

\

$P_{1}(x) = x - \frac{<x, 1>}{<1, 1>}1 = x - 0 = x$

\

$\hspace{5em}<x, 1> = \int_{-1}^{1}xdx = \left[\frac{x^{2}}{2}\right]_{-1}^{1} = \frac{1}{2} - \frac{1}{2} = 0$

\

$\hspace{5em}\therefore \boxed{P_{1}(x) = x}$

\

$P_{2}(x) = x^{2} - \frac{<x^{2}, 1>}{<1, 1>}1 - \frac{<x^{2}, x>}{<x, x>}x = x^{2} - \frac{\frac{2}{3}}{2} = x^{2} - \frac{1}{3}$

\

$\hspace{5em}<x^{2}, 1> = \int_{-1}^{1}x^{2}dx = \left[\frac{x^{3}}{3}\right]_{-1}^{1} = \frac{1}{3} + \frac{1}{3} = \frac{2}{3}$

\

$\hspace{5em}<1, 1> = \int_{-1}^{1}1dx = 2$

\

$\hspace{5em}<x^{2}, x> = \int_{-1}^{1}x^{3}dx = \left[\frac{x^{4}}{4}\right]_{-1}^{1} = 0$

\

$\hspace{5em}\therefore \boxed{P_{2}(x) = x^{2} - \frac{1}{3}}$

$P_{3}(x) = x^{3} - \frac{<x^{3}, 1>}{<1, 1>}1 - \frac{<x^{3}, x>}{<x, x>}x - \frac{<x^{3}, x^{2}>}{<x^{2}, x^{2}>}x^{2} = x^{3} - 0 - \frac{\frac{2}{5}}{\frac{2}{3}}x - 0 = x^{3} - \frac{3}{5}x$

\

$\hspace{5em}<x^{3}, x> = \int_{-1}^{1}x^{4}dx = \left[\frac{x^{5}}{5}\right]_{-1}^{1} = \frac{1}{5} + \frac{1}{5} = \frac{2}{5}$

\

$\hspace{5em}<x, x> = \int_{-1}^{1}x^{2}dx = \left[\frac{x^{3}}{3}\right]_{-1}^{1} = \frac{2}{3}$

\

$\hspace{5em} \therefore \boxed{P_{3}(x) = x^{3} - \frac{3}{5}x}$ 


\


Ahora, para normalizar los polinomios, multiplicamos por una constante de normalizacion, llamemosla $C_{n}$

\

$C_{0}^{2}\int_{-1}^{1}P_{0}(x)P_{0}(x)dx = C_{0}^{2}\int_{-1}^{1}1dx = C_{0}^{2}(2) = 1 \xrightarrow{} C_{0} = \sqrt{\frac{1}{2}}$

\

$C_{1}^{2}\int_{-1}^{1}P_{1}(x)P_{1}(x)dx = C_{1}^{2}\int_{-1}^{1}x^{2}dx = C_{1}^{2}\left[\frac{x^{3}}{3}\right]_{-1}^{1} = C_{1}^{2}\left(\frac{2}{3}\right) = 1 \xrightarrow{} C_{1} = \sqrt{\frac{3}{2}}$

\

$C_{2}^{2}\int_{-1}^{1}P_{2}(x)P_{2}(x)dx = C_{2}^{2}\int_{-1}^{1}(x^{2}-\frac{1}{3})^{2}dx = C_{2}^{2}\left[\frac{8}{48}\right]=1 → C_{2} = \frac{3}{2}\sqrt{\frac{5}{2}}$

\

$C_{3}^{2}\int_{-1}^{1}\left(x^{3}-\frac{3}{5}x\right)^{2} = C_{2}^{2}\frac{8}{175} = 1 → C_{3} = \frac{5}{2}\sqrt{\frac{7}{2}}$

\

Multiplicando estas constantes por los polinomios hallados anteriormente se obtiene los siguiententes polinomios de Legendre normalizados:

\

$\boxed{P_{0}(x) = \sqrt{\frac{1}{2}}}$

\

$\boxed{P_{1}(x) = \sqrt{\frac{3}{2}}x}$

\

$\boxed{P_{2}(x) = \sqrt{\frac{5}{2}}\left(\frac{3x^{2}}{2} - \frac{1}{2}\right)}$

\

$\boxed{P_{3}(x) = \sqrt{\frac{7}{2}}\left(\frac{5}{2}x^{3}-\frac{3}{2}x\right)}$


```python
#Graficas de los primeros cuatro polinomios de Legendre (P_0(x),P_1(x),P_2(x),P_3(x)) normalizados

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import legendre

N = 1000
xval = np.linspace(-1,1,N)

def elilio(x,n):
  leg = legendre(n)
  P_n = leg(x)
  return P_n

P0 = np.sqrt(1/2)*(np.array(elilio(xval,0)))
P1 = np.sqrt(3/2)*(np.array(elilio(xval,1)))
P2 = np.sqrt(5/2)*(np.array(elilio(xval,2)))
P3 = np.sqrt(7/2)*(np.array(elilio(xval,3)))

plt.plot(xval,P0,'g',label = "n=0")
plt.plot(xval,P1,'b',label = "n=1")
plt.plot(xval,P2,'r',label = "n=2")
plt.plot(xval,P3,'c',label = "n=3")

plt.title ("First 4 P_n(x)")
plt.grid()
plt.legend()
plt.show()
```

donde la funcion debe tomar la siguiente forma

$a_{0} = \frac{1}{L}\int_{0}^{L}x^{2}dx = \frac{L^{2}}{3}$

$\boxed{f(x) = x^{2} = \frac{L^{2}}{3}+\frac{4L^{2}}{\pi^{2}}\sum^{\infty}_{n=1}\frac{(-1)^{n+1}}{n^{2}}cos(\frac{n\pi x}{L})}$

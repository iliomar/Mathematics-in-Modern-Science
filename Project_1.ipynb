A simple vista se puede ver que hay $3$ puntos de equilibrio. Estos son, $(0, 0)$, $\approx (3.2, 1.51)$ y $\approx (-3.2, -1.51)$. El punto de equilibrio $(0,0)$ es inestable, pues algunas condiciones divergen del punto de equilibrio, este punto de equilibrio es una fuente. Por otro lado, los puntos de equilibrio $\approx (3.2, 1.51)$ y $\approx (-3.2, -1.51)$ son asintoticamente estables, ya que las condiciones iniciales cercanas convergen al punto de equilibrio, este punto de equilibrio es un atractor.

```python
# import necessary libraries
import matplotlib.pyplot as plt
from math import *
# show plots in notebook
% matplotlib inline

# define system in terms of separated differential equations
def f(x,y):
    return y - 0.5*x
def g(x,y):
    return sin(x)

# initialize lists containing values
x = []
y = []

#iv1, iv2 = initial values, dt = timestep, time = range
def sys(iv1, iv2, dt, time):
    # initial values:
    x.append(iv1)
    y.append(iv2)
    #z.append(iv3)
    # compute and fill lists
    for i in range(time):
        x.append(x[i] + (f(x[i],y[i])) * dt)
        y.append(y[i] + (g(x[i],y[i])) * dt)
        #z.append(z[i] + (h(x[i],y[i],z[i])) * dt)
    return x, y

sys(5, 2, 0.01, 1000)

#plot
fig = plt.figure(figsize=(15,5))
fig.subplots_adjust(wspace = 0.5, hspace = 0.3)
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.plot(x, 'r-', label='predator')
ax1.plot(y, 'b-', label='prey')
#ax1.plot(z, 'g-', label='prey')
ax1.set_title("Dynamics in time")
ax1.set_xlabel("time")
ax1.grid()
ax1.legend(loc='best')

ax2.plot(x, y, color="blue")
ax2.set_xlabel("x")
ax2.set_ylabel("y")  
ax2.set_title("Phase space")
ax2.grid()
```

Para verificar algebraicamente los puntos de equilibrio, simplemente se iguala a cero ambas ecuaciones diferenciales.

\

$f(x, y) = \frac{dx}{dt} = y - 0.5x = 0\\
g(x, y) = \frac{dy}{dt} = sin(x) = 0$

\

\

Para que $sin(x) = 0$ $x$ debe ser $0$ o $\pm\pi$. Por lo tanto, se tiene que $x = 0, \pi$

Sustituyendo la x se obtienen los siguientes resultados,

$y - 0.5(0) = 0 \longrightarrow y = 0\\
y - 0.5(\pi) = y-1.57 =0 \longrightarrow y = 1.57\\
y - 0.5(-\pi) = y+1.57 =0 \longrightarrow y = -1.57$

Finalmente, se concluye que los puntos correspondientes son:

$(0, 0)$, $(\pi, 1.57)$ y $(-\pi, -1.57)$

(c) Para linearizar alrededor de cada punto de equilibrio calculamos la matriz jacobiana y se evalua en los respectivos puntos.

\

$J(x, y) = \left(\begin{array}{rr}
\frac{\partial f(x, y)}{\partial x}& \frac{\partial f(x, y)}{\partial y}\\
\frac{\partial g(x, y)}{\partial x}& \frac{\partial g(x, y)}{\partial y}
\end{array}\right) = \left(\begin{array}{rr}
-0.5 & 1\\
cos(x) & 0
\end{array}\right)$

$J(0, 0) = \left(\begin{array}{rr}
-0.5 & 1\\
cos(0) & 0
\end{array}\right) = \left(\begin{array}{rr}
-0.5 & 1\\
1 & 0
\end{array}\right)$

\

$J(\pi, 1.57) = \left(\begin{array}{rr}
-0.5 & 1\\
cos(\pi) & 0
\end{array}\right) = \left(\begin{array}{rr}
-0.5 & 1\\
-1 & 0
\end{array}\right)$

\

$J(-\pi, -1.57) = \left(\begin{array}{rr}
-0.5 & 1\\
cos(-\pi) & 0
\end{array}\right) = \left(\begin{array}{rr}
-0.5 & 1\\
-1 & 0
\end{array}\right)$

Como se puede observar en el punto de equilibrio $(\pi, 1.57)$ y $(-\pi, -1.57)$, tanto los autovalores como los autovectores son numeros imaginarios. Esto quiere decir que las trayectorias son curvas cerradas que rodean un centro.

```python
def plotVectors(vecs, cols, alpha=1):    
    plt.axvline(x=0, color='#A9A9A9', zorder=0)
    plt.axhline(y=0, color='#A9A9A9', zorder=0)

    for i in range(len(vecs)):
        x = np.concatenate([[0,0],vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=cols[i],
                  alpha=alpha)
```

```python
v = np.array([[-1], [4]])
Xv = X.dot(v)
v_np = [-0.61541221, -0.78820544]

plotVectors([Xv.flatten(), v.flatten(), v_np], cols=['green', 'orange', 'blue'])
plt.ylim(-4, 5)
plt.xlim(-7, 5)
```

En este ejercicio, las soluciones del sistema no lineal y las del linearizado son casi aproximadamente las mismas cerca del origen. Pero, como el sistema linearizado es un centro, cualquier perturbacion puede modificar le comportamiento de las soluciones. Existen dos situaciones en donde el comportamiento de las soluciones cerca de un punto de equilibrio no lineal y su linearizacoin pueden diferir. Cuando el sistema linearizado es un centro y cuando tiene $0$ como autovalor. En cualquier otro caso, el comportamiento de las soluciones de un sistema lineal y no lineal cerca del punto de equilibrio es el mismo para las soluciones de su linearizacion.

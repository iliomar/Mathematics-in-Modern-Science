```python
import numpy as np
import pylab as plt
import math
import random
import sympy as sp
from scipy import optimize as opt

#Definiendo curva: 

#Curva linea
def polinomica(a):
  degree = len(a)
  y = 0
  for i in range(degree):
    y += a[i]*x*(degree-i-1)
  return y

#Curvas con ruidos de 10%

def ruido(y):
  yr = np.copy(y)
  for i in range(len(y)):
    yr[i] = yr[i]+yr[i]*0.2*(random.random()*2-1)
  return yr


x =np.linspace(1,31,30)
al = [100,2,5]
yl = polinomica(al)

#Puntos
p = yl[0:10]
y_ruidol = ruido(p)
```

```python
# Se optimiza y se ajusta la curva a datos usando Tikhonov

# Suma de cuadrados del modelo vs datos
def SSQ1(z):
  sum1 = sum((y_ruidol-p)**2)
  return sum1

# Suma de cuadrados de parámetros
def SSQ2(z):
  sum2=sum(np.array(z)**2)
  return sum2

# Función objetivo, con parámetro de regularización alpha
def obj(z):
  sum3=SSQ1(z)+(alpha**2)*SSQ2(z)
  return sum3
  
data = np.zeros((3+4,3))

# minimizar objetivo
for i in np.arange(-4,3):
  alpha = 10**float(i)
  xxopt = opt.minimize(obj,[90,1,1],method='CG')
  z = [xxopt.x[0],xxopt.x[1],xxopt.x[2]]
  SSQ1v = SSQ1(z)
  SSQ2v = SSQ2(z)
  data[i+4,0], data[i+4,1] , data[i+4,2] = float(alpha),float(SSQ1v),float(SSQ2v)

  # Gráficas
  f = plt.figure()
  f.set_figwidth(20)
  f.set_figheight(8)
  plt.plot(x,yl,label="real")


  ymodel = polinomica(xxopt.x)
  plt.plot(x[0:10],y_ruidol,'or',label="ruido")
  plt.plot(x,ymodel,label="Modelo")
  plt.legend()
  plt.xlabel('x')
  plt.ylabel('y')
  plt.grid()
  plt.title('Alpha ' + str(alpha))
  plt.show()
```

```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt

# Se define rango de datos con ruido
start = 0
stop = 10
increment = 1
x = np.arange(start,stop,increment)

# parámetros del modelo logístico
a = 1000
m = 10
n = 50
tau = 5

# modelo logístico
def model(x, a, m, n, tau):
  y = a * (1+m*np.exp(-x/tau))/(1+n*np.exp(-x/tau))
  return y

# inicializamr el generador aleatorio y crear ruido
np.random.seed()
y = model(x,a,m,n,tau)
for i in x:
  y_noise = np.random.random()
  # Añadir ruido al modelo
  y[i] = y[i] + 0.1*(y[i]*(2*y_noise-1))
```

```python
# Optimizamos, ajustamos curva a datos usando Tikhonov
# Suma de cuadrados del modelo vs datos
def SSQ1(z):
  sum1=0
  y1 = z[0] * (1+z[1]*np.exp(-x/z[3]))/(1+z[2]*np.exp(-x/z[3]))
  for i in range(0,len(y)):
    sum1 = sum1+(y[i]-y1[i])**2
  return sum1

# Suma de cuadrados de parámetros
def SSQ2(z):
  sum2=0*z[0]**2+z[1]**2+z[2]**2+z[3]**2
  return sum2

# Función objectivo
def obj(z):
  sum3=SSQ1(z)+(alpha**2)*SSQ2(z)
  return sum3
  
data = np.zeros((3+4,3))
# minimizar objetivo, se elige método CG, hay muchos métodos
for i in np.arange(-4,3):
  alpha = 10**float(i)
  xxopt = opt.minimize(obj,[100,1,1,1],method='CG')
  z = [xxopt.x[0],xxopt.x[1],xxopt.x[2],xxopt.x[3]]
  SSQ1v = SSQ1(z)
  SSQ2v = SSQ2(z)
  data[i+4,0], data[i+4,1] , data[i+4,2] = float(alpha),float(SSQ1v),float(SSQ2v)

  # Las gráficas
  f = plt.figure()
  f.set_figwidth(20)
  f.set_figheight(8)
  # Graficamos datos con ruido
  plt.plot(x,y, 'or')
  # Graficamos datos, modelo ajustado y modelo teórico
  increment = 1
  xmodeldata = np.arange(start,5*stop,increment)
  ymodel = model(xmodeldata, *xxopt.x)
  ytrue = model(xmodeldata, a,m,n,tau)
  plt.plot(xmodeldata,ymodel)
  plt.plot(xmodeldata,ytrue,'--')
  plt.legend(["Datos con ruido","Modelo ajustado","Modelo"], loc ="lower right")
  plt.xlabel('Día #')
  plt.ylabel('y')
  plt.grid()
  plt.title('Alpha ' + str(alpha))
  print([a,m,n,tau],xxopt.x)
  plt.show()
```

```python
from pylab import scatter,show,xlabel,ylabel
import pylab as plt


n = data[:,0]
fig, ax = plt.subplots()
x = data[:,1]
y = data[:,2]
scatter(x,y)
xlabel("SSQ1")
ylabel("SSQ2")
for i, txt in enumerate(n):
  ax.annotate(txt, (x[i], y[i]))

show()
```

Sabemos que el fin de la regularizacion es evadir grandes cambios en el modelo ajustado debido a cambios pequeños en los datos. EN este caso se puede ver que valores entre 0.01 y 0.2 tienen un buen balance entre SSQ1 y SSQ2 por lo que 0.1 puede ser un buen valor para $\alpha$

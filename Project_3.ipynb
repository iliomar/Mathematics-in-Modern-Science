#Ejercicio 4

Usando Python, tome este modelo logístico: $$f(t)=a\frac{1+me^{-t/\tau}}{1+ne^{-t/\tau}}$$ con los siguientes valores en los parámetros: $a=560$, $m=11$, $n=55$, $\tau=6$, y genere "datos sintéticos" en el tiempo, evaluando la función y añadiendo 10% de "ruido aleatorio". Por ejemplo, $$y_1=f(t_1)+0.1*f(t_1)*(u_1-1),$$ donde $u_1$ es una variable aleatoria uniforme (un número aleatorio) entre 0 y 2. Así, para cada instante en el tiempo en el que queremos un dato *ruidoso*.

Entonces, tome los primeros $n$ datos y ajuste el modelo logístico. Luego repita con los primeros $n+1$ datos. Luego con los primeros $n+2$ datos, etc.

Lea esta [lectura](https://constancecrozier.com/2020/04/16/forecasting-s-curves-is-hard/), que incluye un vídeo, y trate de hacer una animación o vídeo parecido con lo hecho anteriormente. En otras palabras, haga un gif o vídeo o animación del proceso de ir añadiendo datos al modelo, recalculando la curva de mejor ajuste y observando si la predicción a corto, mediano o largo plazo sufre cambios grandes.

OJO: una alternativa para hacer la animación es juntar todas las gráficas de cada modelo como si fueran *frames* de una película y usar un *online gif maker* para combinarlas y hacer la animación.

¿Observa sensitividad a cambios (añadiduras) en los datos? ¿Es útil este proceso para generar proyecciones útiles a corto, mediano o largo plazo?


```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# definimos un range para generar datos con ruido
start = 0
stop = 120
increment = 1
x = np.arange(start,stop,increment)


# parámetros del modelo logístico
a = 560
m = 11
n = 55
tau = 6


# modelo logístico
def model(x, a, m, n, tau):
  y = a * (1+m*np.exp(-x/tau))/(1+n*np.exp(-x/tau))
  return y

# inicilizamos el generador aleatorio y creamos ruido

np.random.seed()
y = model(x,a,m,n,tau)
y_noise =[]
for i in x:
  r = (2*np.random.random()-1)
  # Añadimos ruido al modelo
  y_noise.append(y[i] + 0.1*y[i]*(r))

# Optimizamos,ajustamos curva a datos

for i in range(len(x)-18):  
  popt, pcov = curve_fit(model, x[:18+i], y_noise[:i+18],(500,1,32,7),method='trf')
  a,m,n,tau = popt[0],popt[1],popt[2],popt[3]
  plt.plot(x,y, "k", label="Datos sin ruido")
  plt.plot(x[:i+18],y_noise[:i+18],"gx",label="Datos con ruidos")
  y1 = model(x,a,m,n,tau)
  plt.plot(x,y1, "r", label="Datos de Modelo Ajustado")
  plt.legend()
  plt.xlabel('x')
  plt.ylabel('y')
  plt.grid()
  plt.title('Modelo Logistico')
  plt.show()
```

Deduzca las fórmulas de los coeficientes $m$ y $b$ de la recta de mejor ajuste $y=mx+b$ usando cálculo, es decir, minimizando la suma de cuadrados de los errores cometidos por la curva de ajuste. HINT: Tome la derivada con respecto a los coeficientes y encuentre el punto crítico.

Sabemos que la ecuacion de una recta es y = mx + b. Por lo tanto, para cada punto de la linea se tendra algunos errores, estos errores se pueden escribir de la siguiente manera como $y_{i} = mx_{i}+b+ϵ_{i}$

Despejando para $ϵ_{i}$(el residuo) se obtiene

$ϵ_{i} = y_{i}-mx_{i}-b$


\

\

Ahora, para minimizar las distancias de los puntos de la recta al mismo tiempo se debe de minimzar la suma, esto es;

\

$\hspace{15em}\sum_{i=1}^{n} ϵ_{i}^{2} = \sum_{i = 1}^{n} (y_{i}-mx_{i}-b)^{2}$

\

Se prosigue hayando la derivada respecto a $m$ y $b$

Donde la derivada de $m$ es

$\frac{\partial Tr}{\partial m} = \sum_{i = 1}^{n} (mx_{i}+b-y_{i})^{2}$

\


$\hspace{1.5em} = \sum_{i = 1}^{n} 2(mx_{i}+b-y_{i})$

\

$\hspace{1.5em} = \sum_{i = 1}^{n} x_{i}(mx_{i}+b-y_{i})$

\

$\hspace{1.5em} = \sum_{i = 1}^{n} mx_{i}^{2}+bx_{i}- y_{i}x_{i}$

\

$\hspace{1.5em} = m\sum x_{i}^{2}+b\sum x_{i} - \sum x_{i} y_{i} = 0$

\

$\hspace{1.5em} m\sum x_{i}^{2} + b\sum x_{i} = \sum x_{i}y_{i}$

\


$\boxed{m = \frac{n\sum x_{i}y_{i} - \sum x_{i} \sum y_{i}}{n\sum x^{2}_{i} - (\sum x_{i})^{2}}}$

\

\

Y la derivada de $b$ es

$\frac{\partial}{\partial b}\sum_{i=1}^{n} ϵ_{i}^{2} = \frac{\partial Tr}{\partial b} = \sum_{i = 1}^{n}2(mx_{i}+b-y_{i})$

\

$\hspace{7.5em} = \sum_{i=1}^{n}y_{i}-mx_{i}+b - y_{i} = 0$

\

Sabemos que $\sum_{i=1}^{n} = b+b+...b = nb$, entonces

$\hspace{7.5em} = m\sum x_{i}+nb - \sum y_{i}$

\

$\hspace{7.5em} = m\sum x_{i} + nb = \sum y_{i}$

\

Por lo tanto, se tiene

\

$m\sum x_{i}+nb = \sum y_{i}$

\

$\frac{nb}{n} = \frac{\sum y_{i} - m\sum x_{i}}{n}$

Finalmente se tiene que $b$ es

\

$\boxed{b = \frac{\sum y_{i} - m\sum x_{i}}{n}}$
